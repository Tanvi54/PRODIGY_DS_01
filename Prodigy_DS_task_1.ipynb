```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Load Dataset**

```python
df = pd.read_csv('heart_data.csv')  # heart disease dataset from Kaggle
print(df.head())
print(df.dtypes)
```

**Histogram â€” Age Distribution**

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Count')
plt.show()
```

**Bar Chart â€” Gender Distribution**

```python
df['gender_label'] = df['gender'].map({1: 'Male', 0: 'Female'})
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='gender_label', palette={"Male":'lightblue', "Female":'pink'})
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Age Distribution by Diagnosis**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='diagnosis', bins=20, kde=False, palette={0: 'tomato', 1: 'dodgerblue'}, multiple='stack')
plt.title('Age Distribution by Heart Disease Diagnosis')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Diagnosis', labels=['No Disease', 'Heart Disease'])
plt.grid(True)
plt.show()
```

**Bar Chart â€” Diagnosis Count**

```python
sns.countplot(x='diagnosis', data=df, palette='coolwarm')
plt.title('Heart Disease Diagnosis Distribution')
plt.xlabel('Diagnosis (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

**Bar Plot â€” Cholesterol by Diagnosis**

```python
sns.barplot(x='diagnosis', y='cholesterol', data=df, palette='Set2')
plt.title('Cholesterol Levels by Diagnosis')
plt.xlabel('Diagnosis (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol')
plt.show()
```

**Histogram â€” Resting Blood Pressure**

```python
sns.histplot(df['rest_bp'], bins=25, kde=False, color='teal')
plt.title('Distribution of Resting Blood Pressure')
plt.xlabel('Resting BP')
plt.ylabel('Frequency')
plt.show()
```
